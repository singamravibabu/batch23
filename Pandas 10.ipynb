# IMPORTS AND EXPORTS

```python
# import libaries
import numpy as np
import pandas as pd
```

```python
# import json file
nobel = pd.read_json("./data/nobel.json")
nobel.head(5)
```

```python
# get the first row as an item using loc method
chemistry_2019 = nobel.loc[0, "prizes"]
chemistry_2019
```

```python
pd.json_normalize(data=chemistry_2019)
```

```python
pd.json_normalize(data=chemistry_2019, record_path="laureates")
```

```python
pd.json_normalize(
    data=chemistry_2019, 
    record_path="laureates",
    meta=["year", "category"])
```
