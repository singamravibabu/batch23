# OBJECT-ORIENTED PROGRAMMING

- In the real-world, we deal with objects, both tangible and intangible.
- Objects such as an employee, student, citizen, boss, mother, father, son, book, car, bike, chair, table, insurance, account, etc.

- Every object has two aspects to it: **attributes** and **methods**
- Attributes can be considered as characteristics or features.
- Methods can be considered as functionality or actions.

### Attributes (usually variables)
#### Examples:
- Human: (object)
    - human.name = "Shyam"        # name is an attribute
    - human.age = 21
    - human.weight = 70
    - human.height = 168
    - human.nationality = "Indian"
    - human.ethincty = "Asian"
    - human.gender = "Male"
    ...
- Book: (object)
    - book.title = "Attitude"
    - book.author = "John C Maxwell"
    - book.publisher = "Wiley"
    - book.dop = "20/4/1999"
    - book.rating = 5
    - book.pages = 144
    - book.chapters = 12
    ...
- Attributes are varibales that store information (data) about the object
- Access an attribute: `object.attribute`
- Set value for a attribute: `object.attribute = value`

### Methods (usually functions)
- Methods are actions performed by an object or performed on object
#### Examples
- Human: (object)
    - human.speak(topic, lang, occassion, ...)
    - human.walk(type, duration, time, direction)
    - human.run(speed, purpose, duration)
    - human.read(subject, mode, purpose)
    - human.think(subject, ...)
    - human.sit(place, object)
    - human.jump()
    ...
- Book: (object)
    - book.open()
    - book.bookmark(page, how)
    - book.close()
    - book.note()
    - book.foldpage()
    - book.tearpage()
    ....
- Using methods: `object.method([args])`

### How to represent/create an object
1. Create a `class`
2. Then using the `class`, create an `object`

##### Definitions
- A `class` is a blueprint or template to construct/create an object.
- An `object` is an instance of class

- A class defines attributes (variables) and methods (functions) for an object.
- Using one class we can create several objects.

- To create a class use the following syntax:

```
class ClassName:
    statements
```
