# SEABORN

- Seaborn is a visualization library built on top of matplotlib
- We can easily work with pandas dataframes using seaborn

```python
# Importing the required libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

### A theme for your plots

```python
sns.set_theme(style="darkgrid")
```

### Load sample data

```python
df = sns.load_dataset("tips")
df.head(3)
```

### Scatterplot (Relationship)

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.show()
```

```python
sns.scatterplot(
    data=df, 
    x="total_bill", 
    y="tip",
    hue="sex",
    style="time")
plt.show()
```

### Line Plot

```python
sns.lineplot(data=df, x="size", y="tip")
plt.show()
```

### Histogram (Distribution)

```python
sns.histplot(
    data=df,
    x="total_bill",
    bins=20
    )
plt.show()
```

```python
# histogram + kde (kernel density estimation)
sns.histplot(
    data=df,
    x="total_bill",
    bins=50,
    kde=True
)
plt.show()
```

### Bar Plot (Statistical Summary)

```python
sns.barplot(
    data=df,
    x="day",
    y="total_bill"
)
plt.show()
```

```python
# grouped
sns.barplot(
    data=df,
    x="day",
    y="total_bill",
    hue="sex"
)
plt.show()
```

### Count Plot

```python
sns.countplot(
    data=df,
    x="day"
)
plt.show()
```

### Box Plot

```python
sns.boxplot(
    data=df,
    x="day",
    y="total_bill"
)
plt.show()
```

### Violin Plot

```python
sns.violinplot(
    data=df,
    x="day",
    y="total_bill"
)
plt.show()
```

### Regression Plot

```python
sns.regplot(
    data=df,
    x="total_bill",
    y="tip"
)
```

### Linear Model Plot

```python
sns.lmplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="day"
)
```

### Heatmap (Correlation Matrix)

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Spectral")
plt.show()
```

### Pair Plot (Multivariate)

```python
sns.pairplot(df, hue='day')
plt.show()
```

### Faceting (Multi Plot)

```python
# relplot (Recommended)
sns.relplot(
    data=df,
    x="total_bill",
    y="tip",
    col="time",
    hue="day"
)
```

```python
# catplot
sns.catplot(
    data=df,
    x="day",
    y="total_bill",
    kind="box"
)
plt.show()
```
