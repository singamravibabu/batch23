# SEABORN

- Seaborn is a visualization library built on top of matplotlib
- We can easily work with pandas dataframes using seaborn

```python
# Importing the required libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

### A theme for your plots

```python
sns.set_theme(style="darkgrid")
```

### Load sample data

```python
df = sns.load_dataset("tips")
df.head(3)
```

### Scatterplot (Relationship)

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.show()
```

```python
sns.scatterplot(
    data=df, 
    x="total_bill", 
    y="tip",
    hue="sex",
    style="time")
plt.show()
```

### Line Plot

```python
sns.lineplot(data=df, x="size", y="tip")
plt.show()
```
