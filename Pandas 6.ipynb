# PIVOTING AND RESHAPING DATA

```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv("./data/sales_by_employee.csv", parse_dates=["Date"])
sales
```

```python
# Create a pivot table using pivot_table() method
sales.pivot_table(
    index="Date",
    values="Revenue",
    aggfunc="sum"
)
```

```python
# Create a pivot table using pivot_table() method
sales.pivot_table(
    index="Date",
    values="Revenue",
    aggfunc="mean"
)
```

```python
sales.pivot_table(
    index="Date",
    values="Revenue",
    aggfunc=["sum", "mean", "count", "min", "max"]
)
```

```python
sales.pivot_table(
    index=["Date", "Customer"],
    values="Revenue",
    aggfunc=["sum", "mean"]
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Customer",
    values="Revenue",
    aggfunc="mean"
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Customer",
    values="Revenue",
    aggfunc="mean",
    fill_value = ""
)
```
