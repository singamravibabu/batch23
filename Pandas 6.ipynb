# PIVOTING AND RESHAPING DATA

```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv("./data/sales_by_employee.csv", parse_dates=["Date"])
sales
```

```python
# Create a pivot table using pivot_table() method
sales.pivot_table(
    index="Date",
    values="Revenue",
    aggfunc="sum"
)
```

```python
# Create a pivot table using pivot_table() method
sales.pivot_table(
    index="Date",
    values="Revenue",
    aggfunc="mean"
)
```

```python
sales.pivot_table(
    index="Date",
    values="Revenue",
    aggfunc=["sum", "mean", "count", "min", "max"]
)
```

```python
sales.pivot_table(
    index=["Date", "Customer"],
    values="Revenue",
    aggfunc=["sum", "mean"]
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Customer",
    values="Revenue",
    aggfunc="mean"
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Customer",
    values="Revenue",
    aggfunc="mean",
    fill_value = ""
)
```

```python
sales.pivot_table(
    index="Date",
    columns=["Customer","Name"],
    values="Revenue",
    aggfunc="mean",
    fill_value = ""
)
```

```python
sales.pivot_table(
    index="Date",
    values=["Revenue", "Expenses"],
    aggfunc={"Revenue":"sum", "Expenses":"mean"},
    fill_value = ""
)
```

```python
sales.pivot_table(
    index="Date",
    columns="Name",
    values= "Revenue",
    aggfunc="sum",
    fill_value = "",
    margins = True,
    margins_name = "Total"
)
```

```python
by_name_and_date = sales.pivot_table(
    index="Name",
    columns="Date",
    values="Revenue",
    aggfunc="sum"
)
by_name_and_date
```

```python
# following is stacking the pivot data
stacked = by_name_and_date.stack()
stacked
```

```python
# unstacking the given data
stacked.unstack()
```

### Melting a dataset

```python
video_game_sales = pd.read_csv("./data/video_game_sales.csv")
video_game_sales
```

```python
# melting the data
video_game_sales.melt(
    id_vars = "Name",
    value_vars = ["NA", "EU", "JP", "Other"],
    var_name = "Region",
    value_name = "Sales"
)
```
