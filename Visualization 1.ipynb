# MATPLOTLIB LIBRARY

```python
# IMPORTING REQUIRED LIBRARIES
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate data
x = np.linspace(0, 10, 50)
y = x ** 2
```

### Line Plot (Most Common)

```python
plt.plot(x, y)      # the plot() method plot a line graph
plt.show()          # to see the graph use show() method
```

```python
# adding xlabel, ylabel and title to the graph
plt.plot(x, y)
plt.xlabel("Numbers")       # x-axis label
plt.ylabel("Squares")       # y-axis label
plt.title("x ** 2 = y")     # title
plt.show()
```

### Plotting multiple lines

```python
plt.plot(x, x, label="y = x")   # label parameter to label the line to be displayed in legend
plt.plot(x, x**2, label="y = x ** 2")
plt.plot(x, x**3, label="y = x ** 3")
plt.legend()                # legend is for highliting key values and colors
plt.show()
```

### Line Style and Markers

```python
# linestyle parameter to specify the style of the line drawn
plt.plot(x, linestyle="-.")      # different line styles: -, --, :, -.
plt.show()
```

```python
x1 = np.array([1, 2, 3, 4, 5])
y1 = x1 + 5
# marker to mark the points on the graph
plt.plot(x1, y1, marker='x')    # markers: o, s, ^, *, x
plt.show()
```

### Scatter plot

```python
# scatter plot to show the relationship between x and y
plt.scatter(x, y, marker="s")
plt.show()
```

### The Bar Chart

```python
categories = ["A", "B", "C"]
values = [20, 30, 15]
plt.bar(categories, values)
plt.show()
```

```python
plt.barh(categories, values)
plt.show()
```

### Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.show()
```
