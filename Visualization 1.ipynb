# MATPLOTLIB LIBRARY

```python
# IMPORTING REQUIRED LIBRARIES
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate data
x = np.linspace(0, 10, 50)
y = x ** 2
```

### Line Plot (Most Common)

```python
plt.plot(x, y)      # the plot() method plot a line graph
plt.show()          # to see the graph use show() method
```

```python
# adding xlabel, ylabel and title to the graph
plt.plot(x, y)
plt.xlabel("Numbers")       # x-axis label
plt.ylabel("Squares")       # y-axis label
plt.title("x ** 2 = y")     # title
plt.show()
```

### Plotting multiple lines

```python
plt.plot(x, x, label="y = x")   # label parameter to label the line to be displayed in legend
plt.plot(x, x**2, label="y = x ** 2")
plt.plot(x, x**3, label="y = x ** 3")
plt.legend()                # legend is for highliting key values and colors
plt.show()
```

### Line Style and Markers

```python
# linestyle parameter to specify the style of the line drawn
plt.plot(x, linestyle="-.")      # different line styles: -, --, :, -.
plt.show()
```

```python
x1 = np.array([1, 2, 3, 4, 5])
y1 = x1 + 5
# marker to mark the points on the graph
plt.plot(x1, y1, marker='x')    # markers: o, s, ^, *, x
plt.show()
```

### Scatter plot

```python
# scatter plot to show the relationship between x and y
plt.scatter(x, y, marker="s")
plt.show()
```

### The Bar Chart

```python
categories = ["A", "B", "C"]
values = [20, 30, 15]
plt.bar(categories, values)
plt.show()
```

```python
plt.barh(categories, values)
plt.show()
```

### Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.show()
```

### The Pie Chart

```python
sizes = [48, 50, 65]
labels = ["Maths", "Physics", "Chemistry"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.show()
```

### Figure Size and Resolution

```python
plt.figure(figsize=(8, 5), dpi=100)
plt.plot(x, y)
plt.show()
```

### Axis Limits and Ticks

```python
plt.plot(x, y)
plt.xlim(0, 5)
plt.ylim(0, 30)
plt.xticks([0, 1, 2, 3, 4, 5])
plt.show()
```

### Grid

```python
plt.plot(x, y)
plt.grid(True)      # to turn off grid lines use: plt.grid(False)
plt.show()
```

### Subplots (Multiple plots in one figure)

```python
# 2 rows and 1 column: 2 plots
plt.subplot(2, 1, 1)        # subplot(rows, columns, chart_num)
plt.plot(x, x)
plt.subplot(2, 1, 2)
plt.plot(x, x**2)
plt.show()
```

```python
# 2 rows and 2 columns: 4 plots
plt.subplot(2, 2, 1)
plt.plot(x, x)
plt.subplot(2, 2, 2)
plt.plot(x, x**2)
plt.subplot(2, 2, 3)
plt.plot(x, x**3)
plt.subplot(2, 2, 4)
plt.plot(x, x**4)
plt.show()
```

### Recommended: `fig` and `ax` Style

```python
# The subplots() method returns figure and axes (plural)
fig, ax = plt.subplots()    # fig is set figure object and ax is set axes object
ax.plot(x, y)
ax.set_title("The Chart")
ax.set_xlabel("Values")
ax.set_ylabel("Squares")
plt.show()
```

### Text and Annotation

```python
# adding text to chart
plt.plot(x, y)
plt.text(2, 10, "Hello, World!")
plt.show()
```

```python
plt.plot(x, y)
plt.annotate(
    "Peak",
    xy=(10, 100),
    xytext=(6, 30),
    arrowprops=dict(arrowstyle="->")
)
plt.show()
```

### Log Scale

```python
plt.plot(x, y)
plt.yscale("log")
plt.show()
```

### Save Plots

```python
plt.plot(x, x**2)
plt.savefig("rajkumar_chart.png", dpi=100)
plt.show()
```

### Clear vs Close

```python
plt.cla()       # clear current axes
plt.clf()       # clear current figure
plt.close()     # close the figure
```
