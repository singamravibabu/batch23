```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
housing_prices = pd.read_csv("HousingPricesData.csv")

housing_prices = housing_prices[["Zip", "Price", "Area", "Room"]]

housing_prices["PriceperSqm"] = housing_prices["Price"] / housing_prices["Area"]


# sorting housing prices based on Price value
housing_prices_sorted = housing_prices.sort_values('Price', ascending=False)
```

```python
fig, ax = plt.subplots()

x = housing_prices_sorted['Zip'][0:10]
y = housing_prices_sorted['Price'][0:10]
y1 = housing_prices_sorted['PriceperSqm'][0:10]
```

```python
plt.subplot(1, 2, 1)
plt.bar(x, y)
plt.subplot(1, 2, 2)
plt.bar(x, y1)
plt.show()
```
