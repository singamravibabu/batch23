# Tuples

- Tuple is a collection of items in sequence, just like a list
- Unlike list, tuple is immutable
- We cannot add, set, change, delete, reverse, sort items in a tuple
- Syntax: `(item1, item2, item3, item4, item5, ...)`

```python
# Create a tuple of six student names
students = ('naveen', 'uday', 'praveen', 'srikanth', 'anand', 'santosh')
```

```python
# students.append('jagadeesh')  # This will raise an AttributeError because tuples are immutable
```

```python
# accessing items from a tuple is similar to lists using indexes
students[2]
```

```python
# unpacking a tuple, this also works for lists
s1, s2, s3, s4, s5, s6 = students
```
