# WORKING WITH DICTIONARIES

- A *dictionary* is a collection of items
- Unlike list and tuples, dictionaries are *unordered* collection of items
- Because a dictionary is unorded so no indexes for the items
- Each item is a dictionary has *key* and *value* pair
- Syntax:

```
{key1:value1, key2:value2, key3:value3, key4:value4, ...}
```

- In a dictionary, a key can be any immutable object type
- There are no duplicate keys in a dictionary, but there can be duplicate values

````python
# create a country with its code as a dictionary
countries = {"IN":"India", "US":"United States", "CA":"Canada", 
    "UK":"United Kingdom", "SA":"South Africa", "SL":"Srilanka"}
````

````python
# printing a dictionary
print(countries)
````

````python
# create an empty dictionary
catalog = {}
type(catalog)
````

### Accessing item of a dictionary
- To access an item of a dictionary use *key* in the place of *index*
- Syntax:

```
dict_name[key]
```

````python
countries["IN"]
````

````python
countries["SL"]
````

### Changing a value of a time in a dictionary
- Syntax:

```
dict_name[key] = newValue
```

````python
countries["SL"] = "Special Location"
countries
````

### Adding a new item (key/value) pair
- Syntax:

```
dict_name[newKey] = value
```

````python
print(countries)
````

````python
countries["HK"] = "HongKong"
print(countries)
````

````python
countries["IR"] = "Ireland"
print(countries)
````

````python
code = input("Enter a country code")
'''
if code == "IN":
    print("India")
elif code == "IR":
    print("Ireland")
'''
# alternatively
print(countries[code])
````

### Checking whether a key is in dictionary or not
- Syntax:

```
key in dict_name
```

- If the *key* is in dictionary it returns *True*; otherwise *False*.

````python
"IN" in countries
````

````python
code = "UK"
if code in countries:
    print("The country is United Kingdom")
````
