### Sorting a Series object

```python
google = pd.read_csv(
    "./data/google_stocks.csv", 
    parse_dates=["Date"], 
    index_col="Date"
).squeeze()
```

```python
google
```

#### The `sort_values()` method

```python
google.sort_values()    # by default, sort_values() sorts in ascending order
```

```python
# sort_values() method has a 'ascending' parameter to specify sort order
google.sort_values(ascending=False)
```

```python
# battles series
battles.sort_values()   # by default, NaN values are sorted to the end
```

```python
battles.sort_values(ascending=False)   # by default, NaN values are sorted to the end
```

```python
# handling NaN value position in sorting
battles.sort_values(na_position="last")   # this is default
```

```python
battles.sort_values(na_position="first")
```

##### The `dropna()` method to drop the missing values

```python
battles.dropna().sort_values()
```

#### The `sort_index()` for sorting index value

```python
pokemon = pd.read_csv(
    "./data/pokemon.csv", 
    index_col="Pokemon"
).squeeze()
pokemon
```

```python
pokemon.sort_index()    # pokemon.sort_index(ascending=True)
pokemon.sort_index(ascending=False)
```

#### Retrieving the Smallest and Largest values with the nsmallest and nlargest methods

```python
# the following two statements are same
google.nlargest(5)
google.nlargest(n=5)
```

```python
google.nlargest(10)
```

```python
# The following two statements are equivalent
google.nsmallest(4)
google.nsmallest(n=4)
```

### Overwriting a Series using `inplace` parameter

```python
google_copy = google.sort_values()
google_copy.sort_values(inplace=True)
google_copy
```

### The `apply` method
- This can be used to apply the methods

```python
funcs = [min, max, len]
for current_func in funcs:
    print(current_func(google))
```

```python
# two ways to use apply method
google.apply(round)
google.apply(func=round)
```

```python
def single_or_multi(pokemon_type):
    if "/" in pokemon_type:
        return "Multi"
    return "Single"
pokemon.apply(single_or_multi)
```
